```python
from diffusers import DiffusionPipeline

pipeline = DiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4")
```

```python
import torch

# Initialize a prompt
prompt = "polar bear on an iceberg"

# Pass the prompt in the pipeline
pipeline(prompt).images[0]
```
